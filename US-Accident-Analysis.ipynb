### Start Time

```python
df.Start_Time
```

```python
df.Start_Time[0] # The dates are in string format so we have to convert it into Timestamp 
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time) # convert to timestamp type and override the original column
df.Start_Time
```

- Figure out how to show percentage

```python
sns.distplot(df.Start_Time.dt.hour, bins = 24, kde = False, norm_hist = True) #we put 24 bins that we know 24hr in a day
```

- A high percentage of accidents occur between 6 am to 9 am. (Probably people in a hurry to get to work)
- Next highest percentage is 1 pm to 5 pm (Probably over-speeding, rash driving, crossing at wrong places etc.)

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins = 7, kde = False, norm_hist = True)
```

- Is the distribution of accidents by hour the same on weekends as on weekdays.

```python
sunday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sunday_start_time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

- While Monday follows the general 24-hour pattern- two peaks coinciding with work rush
- On Sundays, the peak occures between 10 am and 5pm, unlike weekdays

```python
sns.distplot(df.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

- This could possibly be due to summer break and then the new school term around August-September. But does this hold for all the years under consideration? 
- The highest peek occures in the month of December due to Adverse weather condition in winter.

```python
df_2016 = df[df.Start_Time.dt.year == 2016]
sns.distplot(df_2016.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

- Clearly, there is data missing for the month of January (data was collected from February 2016)
- Much data is missing for 2016. Maybe even 2017

```python
df_2019 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df_2019.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

- 2019 does not show a consistent increase in accidents over the months. So, the overall picture might be misleading due to the significant numbers missing. It could also be attributed to data mismatch if the sources of the collection are varied. This would need further investigation.

## Summary and Conclusion

### Insights:
1. The top 5 states by accidents include populous ones like California, Florida, and Texas.
2. The number of accidents per city decreases exponentially
3. Less than 5% (4.24%) of cities have more than 1000 yearly accidents.
4. Over 1100 cities have reported just one accident 4years needs further investigation.
5. The majority of them have witnessed between 10â€“100 accidents during the period.
6. Less than 100 cities (out of 11,682) reported more than 1000 accidents during the period between February 2016 and December 2020.
7. The hypothesis that weekdays see more accidents during morning and evening rush hours is corroborated by the data. Weekends have the peak during the afternoon.
8. The year-wise analysis seems to show missing data about certain months. So nothing much can be conclusively said about the overall increasing trend in the latter half of the year.
9. The weather condition seem to be â€œFairâ€,â€œCloudyâ€ and â€œMostly Cloudyâ€in most cases. The assumption of a 10mile visibility also doesnâ€™t conclusively explain the accidents due to these factors.
10. Most accidents happened away from the main crossing where there was inadequate or absence of traffic management infrastructures like traffic signaling systems and speed bumps.

### Conclusion
>In conclusion, it can be said that while the percentage of accidents is low for most cities, it is important to hammer the point that this analysis points to the following-

- The majority of such accidents seemed to have happened due to the absence of proper traffic management mechanisms like traffic signals and bumps- particularly in areas away from the main crossings.
